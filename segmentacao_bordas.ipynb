```python
import cv2
from matplotlib import pyplot as plt
import numpy as np

import requests
url = "https://64.media.tumblr.com/5b7f2bc67489cebc1ba17dc3af8acbf4/891c15fd63626fdc-ed/s1280x1920/957bc75e6ba3ba0c75416acd07abf284ee46974f.jpg"
r = requests.get(url)
with open ('test.jpg', 'wb') as f:
  f.write(r.content)
```

```python
img = cv2.imread('test.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img = cv2.GaussianBlur(gray,(3,3),0)
```

```python
sobelx = cv2.Sobel(src=img, ddepth=cv2.CV_64F, dx=1, dy=0, ksize=5)
sobely = cv2.Sobel(src=img, ddepth=cv2.CV_64F, dx=0, dy=1, ksize=5)
sobelxy = cv2.Sobel(src=img, ddepth=cv2.CV_64F, dx=1, dy=1, ksize=5)

plt.figure(figsize=(18, 19))
plt.subplot(221)
plt.imshow(img, cmap='gray')
plt.title('Original')
plt.axis('off')

plt.subplot(222)
plt.imshow(sobelxy, cmap='gray')
plt.title('Sobel X Y')
plt.axis('off')

plt.subplot(223)
plt.imshow(sobelx, cmap='gray')
plt.title('Sobel X')
plt.axis('off')
```
